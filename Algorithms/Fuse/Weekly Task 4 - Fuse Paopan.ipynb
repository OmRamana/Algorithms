```python
#creating a Query URL and defining it as measurement_url

measurement_url = ('https://docs.openaq.org/v2/measurements?'
          'city[]=Beijing'
          '&city[]=Addis%Ababa'
          '&parameter=pm10')
#defining a url to find information for Particulate matter less than 10 micrometers in diameter in Beijing and Addis Ababa as measurement_url

response = requests.get(measurement_url) 
#defining the request as response
#using the library called requests that was imported, then use .get() function to make a GET request to the measurement_url

response #print response for finding the information of particulate matter in the areas

measurement=response.json() 
#parse response of the GET request into JSON format to later view as raw JSON data and save that JSON info as measurement

pp = pprint.PrettyPrinter(indent=1)
#defining parameters as pp, then use PrettyPrinter instance to understand the data by 1 indent parameter

pp.pprint(measurement)
#print particulate matter data returned from the GET request as JSON with pprint function
```
